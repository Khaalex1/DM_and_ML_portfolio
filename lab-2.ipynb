# Data Exploration and Visualisation

The goal of this lab is to explore the chosen Brain MRIs Dataset, i.e. extract some statistical characteristics from it.
As this dataset is made of images, this lab is mainly about visualizing the different types of MRIs and chracterizing the classes' repartition.

## Importing Libraries

```python
import os
import cv2
import random
import numpy as np 
import pandas as pd 
import seaborn as sns

# Scikit-Learn â‰¥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"

# To plot pretty figures
%matplotlib inline
import matplotlib as mlp
import matplotlib.pyplot as plt
```

## Data Loading 

This step loads the images contained in the training dataset.

```python
labels = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']
```

```python
train_img = [] #contains the images used for training the model
train_labels = [] #label of each image in x_train 
PATH = 'archive/Training'

for label in labels:
    img_dir = os.path.join(PATH, label)
    for img_file in os.listdir(img_dir):
        img = cv2.imread(f'{img_dir}/{img_file}')
        train_img.append(img)
        train_labels.append(label)
#train_img = np.array(train_img)
#train_labels = np.array(train_labels)
```

## Sample images 

Here we visualize the first 5 images and check that the images are loaded properly

```python
count = 0

while count < 5:
    i = random.randrange(len(train_img))
    img = train_img[i]
    plt.figure()
    plt.imshow(img)
    plt.title(train_labels[i])
    plt.show()
    count += 1
```

## Resizing the images
The images above are of different sizes of which some of them are very large. We need to resize all the images to a smaller size to help speed up the training process of the models we would be working with. Also, this would make sure that all the images are of the same size which is what a model would expect. Moreover, the current images have 3 channels : R,G and B. However, as we can see below, the perception of the MRIs is still in grayscale (in other terms, the three channels are equal). Therefore, we preprocess the images by converting them in grayscale (only 1 channel), which will enable us to deal with matrices (/2D arrays).

```python
new_size = (255, 255)

for i in range(len(train_img)):
    img = train_img[i] 
    train_img[i] = cv2.resize(img, new_size)
    train_img[i] = cv2.cvtColor(train_img[i], cv2.COLOR_RGB2GRAY)/255
    if i<3:
        fig, ax = plt.subplots(1,2,figsize=(10,10))
        ax[0].imshow(img)
        ax[0].set_title("original RGB image")
        ax[1].imshow(train_img[i], cmap = 'gray')
        ax[1].set_title("Grayscale image")
        plt.show()
        print("Grayscale shape :", train_img[i].shape)
        print("Grayscale max value :",train_img[i].max())
```

```python
train_img = np.stack(train_img)
train_labels = np.stack(train_labels)

print("train_img shape : ", train_img.shape)
print("train_labels shape : ", train_labels.shape)
```

The preprocessing has been successful : the dataset has indeed 2870 images, of shape 255 by 255 pixels.

## Show sample image from each class of tumor

```python
k =0
fig, ax = plt.subplots(1,4,figsize=(20,20))
fig.text(s="Sample Image From Each Class of Tumor", 
         size=20, 
         fontweight='bold',
         fontname='monospace',
         y=0.62,
         x=0.4,
         alpha=0.8
        )

for i in labels:
    j = 0
    while True:
        if train_labels[j] == i:
            sh = ax[k].imshow(train_img[j], cmap = 'gray')
            fig.colorbar(sh, ax=ax[k], shrink=0.2)
            ax[k].set_title(train_labels[j])
            k += 1
            break
        j += 1
            
            
```

## Pie Chart

The Pie chart shows classes of brain tumors in proportion to each other. 

```python
# Get the count of images for each class of tumor

train_dict = {}
class_count = []
tumor_classes = []

for tumor_class in train_labels:
       
    if train_dict.get(tumor_class):
        train_dict[tumor_class] += 1
    else:
        train_dict[tumor_class] = 1

for tumor_class, count in train_dict.items():
    tumor_classes.append(tumor_class)
    class_count.append(count)
    print(str(tumor_class) + " contains: " + str(count) + " images" )
    

```

```python
# Plot the pie chart

plt.pie(class_count, startangle=90,explode=[0.1,0.1,0.1,0.2],autopct='%0.2f%%',
        labels=tumor_classes,radius=3)
plt.show()
```
